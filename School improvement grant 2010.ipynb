```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/userssharedsdfschoolimprovement2010grants.csv')
```

```python
map(lambda x: x, df['Location'])
```

```python
g=df['2010/11/Award Amount']
```

```python
y=0.0
newg = g.apply((lambda x: str(x)[1:]))
```

```python
df['Grant'] =newg
```

```python
grant_copy = df['Grant'].copy()
df['Grant'] = df['Grant'].map((lambda x: np.nan if x=='an' else np.float64(x)))
```

```python
states = df.groupby('State')
df['Grant'].sort(inplace=False)[-20:].sum()
```

```python
state_grant = []
for each in states:
    each[1]['Grant'].astype(np.float64)
    g=each[1]['Grant']
    state_grant.append((each[0],g.sum()))
```

```python
sg=pd.Series(dict(state_grant))
```

```python
sg.sort(ascending=False)
```

```python
#Gross grant amount for schools for all states in US
sg.plot(kind='bar')
```

```python
#Gross grant amount for schools for top 20 states
sg[:20].plot(kind='bar')
```
